```python
## PCA solo con Robust

from sklearn.decomposition import PCA

pca2 = PCA(2)
projected = pca2.fit_transform(reviewRS)
sns.scatterplot(x=projected[:,0], y = projected[:,1], hue = products2.brand)
plt.xlabel('component 1')
plt.ylabel('component2')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.show()
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

## kmeans

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3, random_state=42)
kmeans.fit(reviewRSdummi)
kpredict = kmeans.predict(reviewRSdummi)
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(reviewRSdummi, kpredict)
```

```python
for i in range(2,48):
    kmeans = KMeans(n_clusters = i, random_state=42)
    kmeans.fit(reviewRSdummi)
    kpredict = kmeans.predict(reviewRSdummi)
    sil = silhouette_score(reviewRSdummi, kpredict)
    print('silhoute', round(sil,3), 'num_clusters', i, 'sumOfSquares', round(kmeans.inertia_, 3))
```
